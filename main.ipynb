```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

Hello

```python
def detect_objects(img):
    # Convert BGR to HSV
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Define range of red color in HSV
    lower_red = np.array([0,55,10])
    upper_red = np.array([10,255,255])

    # Define range of pink color in HSV
    lower_pink = np.array([160,100,100])
    upper_pink = np.array([179,255,255])

    # Create masks
    red_mask = cv2.inRange(img_hsv, lower_red, upper_red)
    pink_mask = cv2.inRange(img_hsv, lower_pink, upper_pink)
    mask = red_mask + pink_mask

    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(img,img, mask= mask)
    
    return res
```

Hello

```python
if __name__ == "__main__":
    folder_path = "./original_images/"
    for filename in os.listdir(folder_path):
        img = cv2.imread(os.path.join(folder_path, filename))
        if img is not None:
            res = detect_objects(img)
            cv2.imwrite("./annotated_images/" + filename, res)
            # show the real image and the detected objects
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.imshow(img)
            plt.show()
            res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
            plt.imshow(res)
            plt.show()
```
